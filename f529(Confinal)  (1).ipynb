```python
# Step 1: Create the box plot and find outlier
plt.figure(figsize=(8, 6))
plt.boxplot(df['Subscribers'],vert=False)
plt.title('Box Plot of Subscribers')
plt.ylabel('Subscribers')
plt.show()
```

```python
Q1 = df['Subscribers'].quantile(0.25)
Q3 = df['Subscribers'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out values that are within the bounds
df_filtered = df[(df['Subscribers'] >= lower_bound) & (df['Subscribers'] <= upper_bound)]

```

Calculating Quartiles and IQR:
- quantile(0.25) calculates the first quartile, which represents the 25th percentile of the 'Subscribers' column.
- quantile(0.75) calculates the third quartile, which represents the 75th percentile.
- IQR = Q3 - Q1 computes the interquartile range, which is the range between the first quartile and third quartile.
Determining Outlier Bounds:
- lower_bound = Q1 - 1.5 * IQR sets the lower bound for outliers as 1.5 times the IQR below the first quartile.
- upper_bound = Q3 + 1.5 * IQR sets the upper bound for outliers as 1.5 times the IQR above the third quartile.
Filtering Outliers:
- df_filtered =filters the original DataFrame (df) to only include rows where the 'Subscribers' column falls within the calculated bounds.

```python
# Step 3: Create a box plot of the filtered data (optional)
plt.figure(figsize=(8, 6))
plt.boxplot(df_filtered['Subscribers'],vert=False)
plt.title('Box Plot of Subscribers (Outliers Removed)')
plt.ylabel('Subscribers')
plt.show()
```

- Remove the symbol Youtuber,Title

```python
#mptystring='': This initializes an empty string to store the characters that are found in sttext.
#for i in sttext: This starts a loop that iterates through each character i in the input sttext.
#if i in finds:: This checks if the character i is present in the finds string.
#If i is in finds, it appends it to the emptystring.
#Finally, the function returns the emptystring which contains the characters from sttext that are in finds.
def symbol(sttext,finds):
    emptystring=''
    for i in sttext:
        if i in finds:
            emptystring=emptystring+i
    return emptystring
            
finds='abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789.,!$\'- '  
df['Youtuber']=df['Youtuber'].apply(symbol,finds=finds)
df['Title']=df['Title'].apply(symbol,finds=finds) 
```

```python
# Remove any trailing/ leading whitespace from values
df['Youtuber'] = df['Youtuber'].str.strip()
df['Title'] = df['Title'].str.strip()
```

```python
# Filter for YouTube channels with 0 video views
remove = df[df['Video Views'] == 0]
remove
```

```python
#Views' is 0, the function returns True, indicating that the row should be removed. If 'Video Views' is not 0, the function returns False, 
#indicating that the row should be kept.
#When axis=1, it applies the function to each row separately (row-wise operation).
#df. This means it checks each row to see if the condition in the remov function is met.
#~ is used to negate the boolean Series returned by df.apply(remov, axis=1). This means that rows where remov returns True will become False, and
#vice versa. In other words, it filters out the rows for which the condition is True.
def remov(row):
    if row['Video Views'] == 0:
        return True
    else:
        return False


df = df[~df.apply(remov, axis=1)]
```

```python
#iloc in pandas is a method used for integer-location based indexing for selection by position.
#It allows you to select rows and columns by their numerical index. 
#df.iloc[:, -1]: This uses the iloc method to select data from the DataFrame df.
#Therefore, df.iloc[:, -1] selects all rows from the last column of the DataFrame, which is typically the target variable
#The result is assigned to the variable y1, indicating that y1 will contain the values of the target variable.
#df.iloc[:, :-1]: Similar to the previous step, this uses iloc to select data from the DataFrame `df`.
#Therefore,df.iloc[:, :-1] selects all rows from all columns except the last one, which are typically the input features.
#The result is assigned to the variable x1, indicating that x1 will contain the values of the input features.
y1=df.iloc[:,-1]
x1=df.iloc[:,:-1]
```

```python
#Categorical Encoding
```

```python
#Xn = pd.get_dummies(x1) is used to perform one-hot encoding on the DataFrame x1 and assign the result to a new DataFrame Xn.
Xn=pd.get_dummies(x1)
Xn
```

```python
# feature scaling
```

```python
#sc.fit_transform(Xn)`: This line applies the standardization transformation to the DataFrame `Xn`.
#sc.fit(Xn) fits the scaler to the data, computing the mean and standard deviation for each feature in `Xn`.
#sc.transform(Xn)` then applies the transformation using the computed mean and standard deviation.
#The result of the transformation is assigned to the variable `xnew`, which will now contain the standardized features.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xnew=sc.fit_transform(Xn)
xnew
```

```python
#Separate testing and traing data
```

```python
#xnew: This is the standardized feature data.
#y1:This is the target variable data.
#test_size=0.3: This parameter specifies that 30% of the data will be used for testing, and the remaining 70% will be used for training.
#random_state=50`: This sets a seed for the random number generator. This ensures that the data is split in the same way every time you run the code, which is important for reproducibility.

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xnew,y1,test_size=0.3,random_state=50)
```

# DecisionTreeClassifier

```python
#criterion='entropy':specifies that the criterion for splitting nodes should be based on entropy. In simpler terms, 
#it uses the concept of entropy (information gain) to decide the best split.
#random_state=14: This sets the seed for the random number generator. It ensures that the results are reproducible. Using the same seed will yield the same results every time you run the code.
#ccp_alpha=0.0: This parameter is used for pruning the tree. A higher value of ccp_alpha leads to more aggressive pruning. 
#In this case, it's set to 0.0, indicating no pruning.
#dt = : This assigns the created DecisionTreeClassifier instance to the variable dt. 
#You can use this variable to train and make predictions with the model.
#Now, you have an instance of the DecisionTreeClassifier ready to be trained on your data. If you want to proceed with training and evaluation, you can use this dt object. If yo
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy',random_state=14,ccp_alpha=0.0)
```

```python
#dt.fit(x_train, y_train) means that you are using the training data to train the Decision Tree Classifier.
#After this step, the model will have learned the patterns in the data and will be ready to make predictions.
dt.fit(x_train,y_train)
```

```python
#y_pred = dt.predict(x_test) means that you are using the trained Decision Tree Classifier to predict the labels for the test data. 
#The predicted labels will be stored in the variable y_pred. You can then compare these predictions with the actual labels (y_test) to evaluate the performance of the model.
y_pred = dt.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

Removing outliers from a dataset is important for several reasons:

1. **Preserving Data Integrity**:
   - Outliers can sometimes indicate errors in data collection or measurement. Removing them helps ensure that the dataset accurately reflects the phenomena being studied.

2. **Improving Model Performance**:
   - Outliers can have a significant impact on the performance of statistical and machine learning models. They can skew results, leading to inaccurate predictions or conclusions.

3. **Avoiding Incorrect Conclusions**:
   - Outliers can distort summary statistics and visualizations, potentially leading to incorrect conclusions about the underlying data.

4. **Enhancing Visualization**:
   - Outliers can make it difficult to visualize the main trends and patterns in the data. Removing them can lead to clearer and more informative visualizations.

5. **Preventing Overfitting**:
   - Outliers can cause some models to overfit to the noise in the data, leading to poor generalization performance on new, unseen data.

6. **Maintaining Assumptions of Statistical Tests**:
   - Many statistical tests and models assume that the data follows a certain distribution or has certain properties. Outliers can violate these assumptions, leading to incorrect results.

7. **Reducing Noise**:
   - Outliers can introduce noise into the data, making it harder to identify meaningful patterns or relationships.

8. **Improving Data Quality**:
   - Outliers can sometimes be the result of data entry errors, measurement inaccuracies, or other data quality issues. Removing them can help improve the overall quality of the dataset.

9. **Better Interpretability**:
   - Outliers can sometimes represent rare events or extreme cases that are not representative of the general trend in the data. Removing them can lead to more interpretable results.

It's important to note that while removing outliers can be beneficial, it should be done carefully and based on a thorough understanding of the data and the context in which it's being used. Additionally, the method used for outlier detection and removal should be chosen based on the specific characteristics of the dataset and the goals of the analysis.
